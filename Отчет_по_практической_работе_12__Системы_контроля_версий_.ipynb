##Задачи работы

1. Установить GIT

Блокнот Google Colab

# Ход работы

1. Решение задачи 1, установить GIT c официального сайта.

2. Решение задачи 2, создать локальный репозиторой, для этого мы создаем любую папку например $project$, затем открываем терминал, переходим в эту папку и пишем команду git init, которая создает локальный репозиторий.

Реализация решения задачи выполнена в пункте "Результат выполнения работы"

3. Решение задачи 3, файл отчета по выполнению практической работы поместить его в созданный локальный репозиторий, для этого скачиваем отчёт с Google Colab и перекидываем его в папку $project$, а затем выполняем команду в терминале git add, которая добавляет файл в локальный репозиторий.

Реализация решения задачи выполнена в пункте "Результат выполнения работы"

4. Решение задачи 4, зафиксировать в отчете все проделанные шаги и создайте версию отчета в GIT, для этого открываем файл отчёта и добавляем шаги, после чего сохраняем и заходим в терминал, где после вызова команды git status мы увидим две версии файла:


GIT_3.png


Чтобы добавить новые изменения достаточно повторить команду git add:

GIT_4.png

Чтобы создать версию контроля GIT заходим консоль и пишем git commit -m "comment", после чего наш файл преобретает версию.

Реализация решения задачи представленна в пункте "Результат выполнения работы"
